Зададим функции для первичного просмотра данных.

```python
# функция для первичного просмотра датасета

def data_look(data):
    # приводим названия всех колонок к змеиному регистру
    data.columns = [x.lower().replace(' ', '_') for x in data.columns]
    print()
    print('\033[1m === Информация по датасету === \033[0m')
    print()
    data.info() # выводим информацию по датасету
    print()
    print('\033[1m === Первые 5 строк данных === \033[0m')
    display(data.head()) # выводим первые 5 строк
    print()
    print('\033[1m === Случайные 5 строк данных === \033[0m')
    display(data.sample(5, random_state=10)) # выводим последние 5 строк  
    print()
    print('\033[1m === Пропуски и уникальные значения по столбцам === \033[0m')
    # выводим данные о пропусках и уникальных значениях по каждому столбцу
    display(pd.concat([data.isna().sum().rename('nan_count'), # количество пропусков
                       data.isna().mean().rename('nan_percent'), # % пропусков
                       data.nunique().rename('unique_values')], # количество уникальных значений
                      axis=1).style.format({'nan_percent' : '{:.2%}'})) 
    print()
    print('\033[1m === Явные дубликаты === \033[0m')
    print()
    print('Количество явных дубликатов:', data.duplicated().sum())
```

```python
# функция для анализа категориальных переменных

def category_look(data, params):
    for param in params:
        print('\033[1m === Параметр:' + param + '=== \033[0m')
        print()
        print('Количество уникальных значений:', len(data[param].unique()))
        print() 
        print(data[param].sort_values().unique())
        print()
        print()
```

### Предобработка данных `game_action`

```python
# выводим дубликаты

game_action[game_action.duplicated(keep=False)]
```

```python
# удаляем дуликаты и проверяем

game_action = game_action.drop_duplicates()
print('Количество оставшихся дубликатов:', game_action.duplicated().sum())
```

```python
# преобразуем данные о времени

game_action['event_datetime'] = pd.to_datetime(game_action['event_datetime'])
print(game_action['event_datetime'].dtype)
```

```python
# выделим в отдельный сотлбец день, чтобы в дальнейшем было удобно группировать по дням

game_action['event_day'] = game_action['event_datetime'].dt.date
game_action['event_day'] = pd.to_datetime(game_action['event_day'])
```

```python
# применяем функцию для просмотра уникальных значений в столбцах с пропущенными значениями

category_look(game_action, ['building_type', 'project_type'])
```

Посмотрим на значения в столбце `event` при отсутствующих знаениях в столбце `building_type`.

```python
game_action[game_action['building_type'].isna()]['event'].unique()
```

Пропущенные значения связаны с завершением первого уровня и разработкой орбитальной станции спутников. Скорее всего игрок не построил никаких зданий, поэтому пропущенные значения в столбце `building_type` заменим на `no_buildings`.

```python
game_action.loc[game_action['building_type'].isna(), 'building_type'] = 'no_buildings'
```

Из правил игры мы знаем, что завершение первого уровня требует от игрока выполнения одного из двух условий:

* Победа над первым врагом;
* Реализация проекта - разработка орбитальной сборки спутников.

Так как в столбце `project_type` мы имеем всего два уникальных значения: это пропуск и `satellite_orbital_assembly`, то скорей всего под пропуском те игроки которые завершили первый уровень благодаря победе над первым врагом и те которые находятся еще в процессе прохождения первого уровня.

Для игроков у который в столбце `event` значение `finished_stage_1` (первый уровень завершён) пропущенные значения в столбце `project_type` заменим на `victory_over_the_enemy`(победа над врагом), для значений `building` (объект построен) и `project` (проект завершён) в столбце `event` пропущенные значения в столбце `project_type` заменим на `not_finish_game`(игра не окончена).

```python
# заполняем пропуски завершивших первый уровень игроков, как победа над врагом

game_action.loc[game_action['project_type'].isna() & (game_action['event'] == 'finished_stage_1'),
                'project_type'] = 'victory_over_the_enemy'
```

```python
# заполняем пропуски не завершивших первый уровень игроков, как игра не окончена

game_action.loc[game_action['project_type'].isna() & (game_action['event'] != 'finished_stage_1'),
                'project_type'] = 'not_finish_game'
```

После обработки пропусков посмотрим на уникальные значения в столбцах `building_type` и `project_type`.

```python
category_look(game_action, ['building_type', 'project_type'])
```

Нам известно, что с одного показа рекламы мы будем получать 0.07 у.е. при постройке каждого здания игроком. Для расчета возможной выручки добавим новый столбец `revenue` в датасет `game_action`.

```python
# функция добавляет столбец с `revenue`

def revenue(data):
    if data == 'building':
        return .07
    else:
        return 0
```

```python
# добавляем столбец с `revenue`

game_action['revenue'] = game_action['event'].apply(revenue)
```

```python
# функция для создания пользовательских профилей

def get_profiles(game_action, ad_costs):

    # сортируем сессии по ID игроков и дате посещения
    # группируем по ID и находим первые значения `session_start` и `source`
    # столбец с временем первого посещения назовём `start_data`
    profiles = (
        game_action.sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg({'event_datetime': [min, max], 'source': 'first'})
        .reset_index()  # возвращаем user_id из индекса
    )
    
    # объеденим разные уровни в названии столбцов
    profiles.columns = list(map("_".join, profiles.columns))
    
    # переименуем столбцы
    profiles = profiles.rename(columns={'user_id_': 'user_id',
                                        'source_first': 'source',
                                        'event_datetime_min': 'start',
                                        'event_datetime_max': 'finish'})
    
    
    
    # определяем дату первого посещения
    # эти данные понадобятся для когортного анализа
    profiles['dt'] = profiles['start'].dt.date
    profiles['day'] = pd.to_datetime(profiles['dt'])

    # считаем количество уникальных пользователей
    # с одинаковыми источником и датой привлечения
    new_users = (
        profiles.groupby(['day', 'source'])
        .agg({'user_id': 'nunique'})
         # столбец с числом пользователей назовём `unique_users`
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()  # возвращаем `day` и `source` из индексов
    )

    # объединяем траты на рекламу и число привлечённых пользователей
    # по дате и каналу привлечения
    adv_costs = ad_costs.merge(new_users, on=['day', 'source'], how='left')
    
    # делим рекламные расходы на число привлечённых пользователей
    # результаты сохраним в столбец `acquisition_cost` (CAC)
    adv_costs['acquisition_cost'] = adv_costs['cost'] / adv_costs['unique_users']
    
    # добавим стоимость привлечения в профили
    profiles = profiles.merge(
        adv_costs[['day', 'source', 'acquisition_cost']],
        on=['day', 'source'],
        how='left',
    )
    
    # отбираем данные только по тем пользователям
    # кто построил здание
    game_action = game_action.query('event == "building"')

    # считаем количество зданий построенных каждым пользователем
    building = (
        game_action.groupby(['user_id'])
        .agg({'event': 'count'})
         # столбец с количеством зданий назовём `building_count`
        .rename(columns={'event': 'building_count'})
        .reset_index()  # возвращаем `user_id` из индексов
    )

    # объединяем профили пользователей с количеством зданий
    # построенных каждым пользователем
    profiles = profiles.merge(building, on=['user_id'], how='left')
    
    return profiles
```

Получим профили пользователей. Для этого вызовем функцию `get_profiles()`, передав ей данные о посещениях и тратах на рекламу:

```python
profiles = get_profiles(game_action, ad_costs)
profiles
```

### Функции для расчёта и визуализации LTV и ROI

```python
# функция для расчёта LTV и ROI

def get_ltv(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        purchases[['user_id', 'event_day', 'revenue']], on='user_id', how='left'
    )
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_day'] - result_raw['start']
    ).dt.days
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='revenue', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # собираем датафрейм с данными пользователей и значениями CAC, 
        # добавляя параметры из dimensions
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()

        # считаем средний CAC по параметрам из dimensions
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost': 'mean'})
            .rename(columns={'acquisition_cost': 'cac'})
        )

        # считаем ROI: делим LTV на CAC
        roi = result.div(cac['cac'], axis=0)

        # удаляем строки с бесконечным ROI
        roi = roi[~roi['cohort_size'].isin([np.inf])]

        # восстанавливаем размеры когорт в таблице ROI
        roi['cohort_size'] = cohort_sizes

        # добавляем CAC в таблицу ROI
        roi['cac'] = cac['cac']

        # в финальной таблице оставляем размеры когорт, CAC
        # и ROI в лайфтаймы, не превышающие горизонт анализа
        roi = roi[['cohort_size', 'cac'] + list(range(horizon_days))]

        # возвращаем таблицы LTV и ROI
        return result, roi

    # получаем таблицы LTV и ROI
    result_grouped, roi_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        roi_grouped  # таблица ROI
        
    )
```

```python
# функция для визуализации LTV и ROI

def plot_ltv_roi(ltv, roi, horizon):

    # задаём сетку отрисовки графиков
    plt.figure(figsize=(15, 10))

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    

    # из таблицы roi исключаем размеры когорт и cac
    roi = roi.drop(columns=['cohort_size', 'cac'])

    # первый график — кривые ltv
    ax1 = plt.subplot(2, 1, 1)
    ltv.T.plot(grid=True, ax=ax1, cmap = 'Accent')
    plt.legend()
    plt.xlabel('Лайфтайм', size=16)
    plt.title('LTV', size=20)

    # второй график — кривые roi
    ax2 = plt.subplot(2, 1, 2)
    roi.T.plot(grid=True, ax=ax2, cmap = 'Accent')
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.legend()
    plt.xlabel('Лайфтайм', size=16)
    plt.title('ROI', size=20)

    plt.tight_layout()
    plt.show()
```

## Проверка гипотез
### Время завершения уровня различается в зависимости способа прохождения

Различие времени завершения уровня в зависимости способа прохождения через реализацию проекта или через победу над первым игроком должно быть проверено через гипотезу о равенстве средних двух генеральных совокупностей:

* H0: время завершения уровня не зависит от способа прохождения (среднее время через реализацию проекта равно среднему времени через победу над первым игроком);
* H1: время завершения уровня зависит от способа прохождения.

Так как мы имеем разные по объему выборки (игроки победившие врага - 5817, игроки реализовавшие проект - 1866) для проверки гипотезы необходимо применить дополнительный параметр `equal_var = False`.

```python
# критический уровень статистической значимости
alpha = .05 

results = st.ttest_ind(
    time_level_victory['time_level'], 
    time_level_project['time_level'],
    equal_var = False)

print("\033[1m Среднее время завершения 1-го уровня: \033[0m")
print("  - Победа над врагом {:.1f} дней.".format(time_level_victory['time_level'].mean()))
print("  - Реализация проекта {:.1f} дней.".format(time_level_project['time_level'].mean()))
print()

print ('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Время завершения 1-го уровня зависит от способа прохождения.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Время завершения 1-го уровня не зависит от способа прохождения.")
```

**Вывод:**

Скорость завершения 1-го уровня зависит от способа его прохождения.

### Зависит ли способ прохождения уровня от источника трафика

* H0: Способ прохождения уровня не зависит от источника трафика
* H1: Способ прохождения уровня зависит от источника трафика

Так как здесь у нас будет сравнение всех источников, то необходимо учесть множественное сравнение. Для этого будем использовать поправку Бонферрони.

```python
# критический уровень статистической значимости

bonferroni_alpha = alpha / 6  # произведено шесть сравнения
```

Напишем функцию для проведения статистического теста.

```python
def stat_test(source_1, source_2, title):
    print('\033[1m' + title + '\033[0m' + '\n')

    results = st.ttest_ind(source_1, source_2)

    print('p-значение: ', results.pvalue)

    if results.pvalue < bonferroni_alpha:
          print("Отвергаем нулевую гипотезу. Способ прохождения 1-го уровня зависит от источника трафика.")
    else:
          print("Не получилось отвергнуть нулевую гипотезу. Способ прохождения 1-го уровня не зависит от источника трафика.")
    
    print('\n')
```

```python
facebook_test = test_data[test_data['source'] == 'facebook']['id_finish']
instagram_test = test_data[test_data['source'] == 'instagram']['id_finish']
yandex_test = test_data[test_data['source'] == 'yandex']['id_finish']
youtube_test = test_data[test_data['source'] == 'youtube']['id_finish']
```

```python
stat_test(facebook_test, instagram_test, 'facebook / instagram')
stat_test(facebook_test, yandex_test, 'facebook / yandex')
stat_test(facebook_test, youtube_test, 'facebook / youtube')
stat_test(instagram_test, yandex_test, 'instagram / yandex')
stat_test(instagram_test, youtube_test, 'instagram / youtube')
stat_test(yandex_test, youtube_test, 'yandex / youtube')
```

**Вывод:**

Способ прохождения уровня не зависит от источника трафика.
